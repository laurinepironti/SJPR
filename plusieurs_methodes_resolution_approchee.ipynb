# Méthode 1 : Approximation par balayage

__La fonction ci-dessous permet d’obtenir des images successives par la fonction $f$ sur l’intervalle $[1;2]$ avec un pas variable $p$.__


```python
def balayage(f,p):
    x=1
    while x<2:
        print("f(",x,")=",f(x))
        x = x+p
    return None
```
