La méthode des moindres carrés, indépendamment élaborée par Legendre et Gauss au début du XIXe siècle, permet de comparer des données expérimentales à un modèle mathématique censé décrire ces données.

La méthode des moindres carrés consiste à minimiser la somme des carrés des écarts entre chaque point du nuage de régression (les points $M_i(x_i ; y_i)$) et son projeté, parallèlement à l'axe des ordonnées, sur la droite de régression (les points $N_i$).

![Droite d'ajustement](https://i.goopics.net/q4c8yb.png)

Cela consiste donc à déterminer la droite rendant la somme : 
$S=(M_1 N_1 )^2+(M_2 N_2 )^2+...+(M_n N_n )^2$ la plus petite possible. On admettra que cette droite existe et est unique

L'objectif de ce devoir est de calculer la valeur de cette somme pour les différentes droites tracées sur votre feuille. En comparant les différentes valeurs de $S$ vous pourrez déterminer quelle est la droite la plus adaptée pour modéliser le phénomène observé.
